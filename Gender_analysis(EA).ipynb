## The percentage trust between the genders for each parliament

A function will be defined to draw the graphs. 

```python
# Defining a function to plot the graph and for user to resize the graph
def plotGraph(politics_party, fig_x, fig_y):
    
    # Make a list for the data
    labels = list(d4.male_data["Countries"])
    women = list(d4.female_data[politics_party])
    men= list(d4.male_data[politics_party])
    
    x = np.arange(len(labels))  
    width = 0.35
    
    # Plot sub-figures and adding paramenters for better visualization
    fig, ax = plt.subplots(figsize=(fig_x, fig_y))
    ax.bar(x - width/2, men, width, label='Men', color='royalblue')
    ax.bar(x + width/2, women, width, label='Women', color='lightpink')
    
    ax.set_ylabel(politics_party)
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation = 90)
    ax.legend()
    
    fig.tight_layout()
    
    return
```

### National Parliament

```python
plotGraph("Trust_Parliament_Percentage", 6 ,4)
plt.title("Figure 2.3: Percentage trust in National Parliament between men and women")
plt.show()
```

It can be observed from the graph that men has a higher percentage trust than women in the National Parliament in most of the country. In Hungary, women has a higher percentage trust in National Parliamenet than the men. The difference between the percentage trust in the National Parliament between men and women are relatively big (except for Finland). 

### European Parliament

```python
plotGraph("Trust_EP_Percentage", 6, 4)
plt.title("Figure 2.4: Percentage trust in European Parliament between men and women")
plt.show()
```

Contrast to FIgure 2.4, women has higher percentage trust in European Parliament than the men in most countries. In Sweden, both gender has the same percentage trust while in Spain, men has a higher percentage trust in the parliament than the women. 

### United Nations

```python
plotGraph("Trust_UN_Percentage", 6, 4)
plt.title("Figure 2.5: Percentage trust in United Nations between men and women")
plt.show()
```

Different to the other 2 graphs, the percentage trust in United Nations amongst men and women are similar for most countries. They have a small difference in the percentage trust. However, in Finland, women are seen to have higher percentage trust than that of men, with a difference of 6%, while in Poland and Belgium, men have a higher percentage trust, with a difference of 3%.

## Conclusion

At a glance, it is perceived that women has higher percentage trust in the parliaments than men. However, when comparing it between the two genders for each parliament, it can be seen that men have a higher percentage trust in National Parliament than women in general. There are still several countries where men has a higher trust in that particular parliament than women, for example Spain and Sweden. Thereafter, it can be concluded that percentage trust in a parliament differ due to the countries that they are in and not due to their gender status.
